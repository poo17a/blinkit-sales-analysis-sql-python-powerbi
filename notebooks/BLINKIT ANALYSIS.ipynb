```python
# Total Sales 
total_sales = df['Sales'].sum()

# Average Sales
avg_sales = df['Sales'].mean()

# Number of Items Sold
no_of_items_sold = df['Sales'].count()

# Average Rating 
avg_rating = df['Rating'].mean()

# Display 

print(f"Total Sales: ${total_sales:,.0f}")
print(f"Average Sales: ${avg_sales:,.0f}")
print(f"Number of Items: {no_of_items_sold:,.0f}")
print(f"Average Rating: {avg_rating:,.1f}")
```

##### **Chart's Requirements**

```python
# Grouped data
sales_by_fat = df.groupby('Item Fat Content')['Sales'].sum()

# Colors for slices
colors = sns.color_palette("colorblind", len(sales_by_fat))

# Create pie chart & capture label objects
plt.figure(figsize=(6,6))
wedges, texts, autotexts = plt.pie(
    sales_by_fat,
    labels=sales_by_fat.index,
    autopct='%1.1f%%',
    startangle=90,
    colors=colors,
    shadow=True,
    wedgeprops={'edgecolor': 'white', 'linewidth': 2.5}
)

# Make labels & percentages bold
for text in texts:
    text.set_fontsize(12)
    text.set_fontweight('bold')

for autotext in autotexts:
    autotext.set_fontsize(12)
    autotext.set_fontweight('bold')

# Title with better styling
plt.title('Sales by Fat Content', fontsize=14, fontweight='bold', pad=15)
plt.axis('equal')  # Keep it circular
plt.tight_layout()
plt.show()
```

```python
# Data
sales_by_type = df.groupby('Item Type')['Sales'].sum().sort_values(ascending=False)

# Colors (Seaborn palette)
colors = sns.color_palette("viridis", len(sales_by_type))

# Create figure
plt.figure(figsize=(12,6))
bars = plt.bar(sales_by_type.index, sales_by_type.values, color=colors, edgecolor='black', linewidth=0.8)

# Rotate x labels
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.yticks(fontsize=10)

# Labels & Title
plt.xlabel('Item Type', fontsize=12, fontweight='bold')
plt.ylabel('Total Sales', fontsize=12, fontweight='bold')
plt.title('Total Sales by Item Type', fontsize=14, fontweight='bold', pad=20)

# Add gridlines
plt.grid(axis='y', linestyle='--', alpha=0.6)

# Remove top & right spines for a cleaner look
#looping through two spines
for spine in ['top','right']:
    plt.gca().spines[spine].set_visible(False)

# Add data labels with background
for bar in bars:
    plt.text(
        bar.get_x() + bar.get_width()/2,
        bar.get_height(),
        f'{bar.get_height():,.0f}',
        ha='center',
        va='bottom',
        fontsize=9,
        fontweight='bold',
        color='black',
        bbox=dict(facecolor='white', alpha=0.8, edgecolor='none', boxstyle='round,pad=0.2')
    )

plt.tight_layout()
plt.show()
```

```python
# Prepare data
grouped = df.groupby(['Outlet Location Type', 'Item Fat Content'])['Sales'].sum().unstack()
grouped = grouped[['Regular', 'Low Fat']]  

# Set style & colors
sns.set_style("whitegrid")
colors = sns.color_palette("Set2", len(grouped.columns))

# Plot
ax = grouped.plot(kind='bar', figsize=(9,6), color=colors, width=0.7)
plt.title('Outlet Tier by Item Fat Content', fontsize=16, fontweight='bold', pad=15)
plt.xlabel('Outlet Location Tier', fontsize=12, fontweight='bold')
plt.ylabel('Total Sales', fontsize=12, fontweight='bold')
plt.legend(title='Item Fat Content', fontsize=10, title_fontsize=11)

# Add data labels
for container in ax.containers:
    ax.bar_label(container, fmt='%.0f', padding=3, fontsize=9)

# Remove top/right spines
sns.despine()

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Prepare data
sales_by_year = df.groupby('Outlet Establishment Year')['Sales'].sum().sort_index()

# Set style
sns.set_style("whitegrid")

# Plot
plt.figure(figsize=(10,6))
plt.plot(
    sales_by_year.index, sales_by_year.values,
    marker='o', markersize=8, linewidth=2.5,
    color=sns.color_palette("deep")[0]
)

# Labels & title
plt.xlabel('Outlet Establishment Year', fontsize=12, fontweight='bold')
plt.ylabel('Total Sales', fontsize=12, fontweight='bold')
plt.title('Total Sales by Outlet Establishment Year', fontsize=16, fontweight='bold', pad=15)

# Data labels
for x, y in zip(sales_by_year.index, sales_by_year.values):
    plt.text(x, y+max(sales_by_year.values)*0.02,  # small offset above point
             f'{y:,.0f}', ha='center', va='bottom', fontsize=9, fontweight='bold',
             bbox=dict(facecolor='white', alpha=0.8, edgecolor='none', boxstyle='round,pad=0.2'))

# Grid & layout
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# Grouped data
sales_by_size = df.groupby('Outlet Size')['Sales'].sum()

# Colors
colors = sns.color_palette("pastel", len(sales_by_size))

# Create figure
plt.figure(figsize=(6,6))
wedges, texts, autotexts = plt.pie(
    sales_by_size,
    labels=sales_by_size.index,
    autopct='%1.1f%%',
    startangle=90,
    colors=colors,
    wedgeprops={'edgecolor': 'white', 'linewidth': 2}
)

# Donut hole
centre_circle = plt.Circle((0,0),0.70,fc='white')
plt.gca().add_artist(centre_circle)

# Style labels & percentages
for text in texts:
    text.set_fontsize(12)
    text.set_fontweight('bold')

for autotext in autotexts:
    autotext.set_fontsize(11)
    autotext.set_fontweight('bold')
    autotext.set_bbox(dict(facecolor='white', edgecolor='none', alpha=0.8, boxstyle='round,pad=0.3'))

# Title
plt.title('Sales by Outlet Size', fontsize=16, fontweight='bold', pad=20)

# Keep circular shape
plt.axis('equal')
plt.tight_layout()
plt.show()
```

```python
# Prepare data
sales_by_location = df.groupby('Outlet Location Type')['Sales'].sum().reset_index()
sales_by_location = sales_by_location.sort_values('Sales', ascending=False)

# Style
sns.set_style("whitegrid")
colors = sns.color_palette("crest", len(sales_by_location))

# Plot with hue = y
plt.figure(figsize=(9,4))
ax = sns.barplot(
    x='Sales', 
    y='Outlet Location Type', 
    data=sales_by_location, 
    palette=colors,
    hue='Outlet Location Type',   # <-- set hue to y
    dodge=False,                  # prevent grouped bars
    legend=False                  # hide legend
)

# Title & labels
plt.title('Total Sales by Outlet Location Type', fontsize=16, fontweight='bold', pad=15)
plt.xlabel('Total Sales', fontsize=12, fontweight='bold')
plt.ylabel('Outlet Location Type', fontsize=12, fontweight='bold')

# Add value labels on bars
for container in ax.containers:
    ax.bar_label(container, fmt='{:,.0f}', padding=3, fontsize=10, fontweight='bold')

# Remove unnecessary borders
sns.despine(left=True, bottom=True)

plt.tight_layout()
plt.show()
```
